## **Question 1 Part 1 Manually fitting for corn and cotton**

```python
import numpy as np

# We first convert data to numpy matrices to manually calculate
X_corn = df_corn_ndvi[['Days_Corn']].to_numpy()
y_corn = df_corn_ndvi[['Corn_NDVI_2013']].to_numpy()

X_corn
```

```python
y_corn
```

```python
X_b = np.c_[np.ones((40, 1)), X_corn] # add x0 = 1 to each instance
X_b
```

```python
w_estimate = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y_corn)     # fit
```

```python
w_estimate
```

Now we can make predictions using w_estimate. There is an example below

```python
# We create an example matrix for testing
x_new = np.array([[195], [234]])
x_new_b = np.c_[np.ones((2, 1)), x_new] # add x0 = 1 to each instance
x_new_b
```

```python
y_predict = x_new_b.dot(w_estimate)          # predict
y_predict
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(X_corn, y_corn, "b.")
plt.plot(x_new, y_predict, "r*")
plt.plot(x_new, y_predict, "g-")
plt.grid()
plt.title("Manual Line Fitting Corn 2013 NDVI From Days")
plt.xlabel("x")
plt.ylabel("y");
```

***At this point we decided write a function to cover everything we did above and added error measurements here to easily compare***

```python
# x and y as string from original df column headers
def homework3(x,y,title):
  df_new = df[[x,y]]
  X_np = df_new[[x]].to_numpy()
  y_np = df_new[[y]].to_numpy()

  # Manual prediction
  X_b = np.c_[np.ones((X_np.size, 1)), X_np] # add x0 = 1 to each instance

  w_estimate = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y_np)

  first = X_np[0][0]
  last = X_np[-1][-1]

  x_new = np.array([ [first],[last] ])

  x_new_b = np.c_[np.ones((2, 1)), x_new] # add x0 = 1 to each instance
  y_predict_mn = x_new_b.dot(w_estimate)

  # Error Measurement for Manual
  mae_sum = 0
  mse_sum = 0
  for y, x in zip(y_np,X_np):
    x_b = np.c_[np.ones((1, 1)), x] # add x0 = 1 to each instance
    prediction = x_b.dot(w_estimate)
    mae_sum += abs(y - prediction)
    mse_sum += (y - prediction)**2
  mse = mse_sum / len(y_np)
  mae = mae_sum / len(y_np)
  print("Mae for Manual is:",mae[0][0])
  print("Mse for Manual is:",mse[0][0])
  print("-----------------------------------------------------------------------")
  


  # Predict with sklearn
  lin_reg = LinearRegression()
  lin_reg.fit(X_np, y_np)
  y_predict_sk = lin_reg.predict(x_new)


  mae_sum = 0
  mse_sum = 0
  for y , x in zip(y_np,X_np):
    prediction = lin_reg.predict(np.array([x]))
    mae_sum += abs(y - prediction)
    mse_sum += (y - prediction)**2
  mse = mse_sum / len(y_np)
  mae = mae_sum / len(y_np)
  print("Mae for SKLearn is:",mae[0][0])
  print("Mse for SKLearn is:",mse[0][0])


  # Plotting
  plt.figure(figsize=(12, 10))

  plt.subplot(211)
  plt.plot(X_np, y_np, "b.")
  plt.plot(x_new, y_predict_mn, "r*")
  plt.plot(x_new, y_predict_mn, "g-")
  plt.grid()
  plt.title(title)
  plt.xlabel("x")
  plt.ylabel("y Manual")

  plt.subplot(212)
  plt.plot(X_np, y_np, "b.")
  plt.plot(x_new, y_predict_sk, "r*")
  plt.plot(x_new, y_predict_sk, "g-")
  plt.grid()
  plt.xlabel("x")
  plt.ylabel("y SK")

  plt.show()
  print("-----------------------------------------------------------------------")
```
